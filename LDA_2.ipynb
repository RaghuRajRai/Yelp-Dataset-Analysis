```python
import pandas as pd
from nltk.corpus import brown
import pyLDAvis
import pyLDAvis.sklearn
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
```

```python
import numpy as np
```

```python
data = pd.read_csv(r"E:\Yelp\Unfiltered Data\YelpZip\Customs\SAMPLE", header=None)
data[4] = data[4].apply(lambda txt: str(txt))

real = data[data[7] == 1]
fake = data[data[7] == -1]
```

```python
vectorizer = CountVectorizer(min_df=10, max_df=0.9, stop_words='english', lowercase=True, token_pattern='[a-zA-Z\-][a-zA-Z\-]{2,}')
lda_model = LatentDirichletAllocation(n_topics=10, max_iter=10, learning_method='online')
```

```python
data_real_vectorized = vectorizer.fit_transform(real[4])
lda_real = lda_model.fit_transform(data_real_vectorized)

pyLDAvis.enable_notebook()
panel = pyLDAvis.sklearn.prepare(lda_model, data_real_vectorized, vectorizer, mds='tsne')
panel
```

```python
data_fake_vectorized = vectorizer.fit_transform(fake[4])
lda_fake = lda_model.fit_transform(data_fake_vectorized)

pyLDAvis.enable_notebook()
panel = pyLDAvis.sklearn.prepare(lda_model, data_fake_vectorized, vectorizer, mds='tsne')
panel
```
